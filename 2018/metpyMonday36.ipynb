## 36 - Matplotlib subplots

[Youtube](https://www.youtube.com/watch?v=JzqATjzogFs)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(100)
y1 = np.random.random(100)
y2 = np.random.random(100)
y3 = np.random.random(100)
```

```python
fig = plt.figure(figsize = (10,8))
ax = plt.subplot(1, 1, 1)
ax.plot(x, y1, x, y2, x, y3)
```

```python
fig = plt.figure(figsize = (10,8))
ax1 = plt.subplot(3, 1, 1)
ax2 = plt.subplot(3, 1, 2)
ax3 = plt.subplot(3, 1, 3)

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
```

```python
fig = plt.figure(figsize = (10,8))
ax1 = plt.subplot(3, 3, 1)
ax2 = plt.subplot(3, 3, 2)
ax3 = plt.subplot(3, 3, 3)

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
```

```python
fig = plt.figure(figsize = (10,8))
ax1 = plt.subplot(3, 3, 1)
ax2 = plt.subplot(3, 3, 5)
ax3 = plt.subplot(3, 3, 9)

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
```
