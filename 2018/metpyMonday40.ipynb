```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(100)
y = np.sqrt(x)
```

```python
fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y, linewidth = 3)

ax.text(70, 9, r'$y = \sqrt{x}$', fontsize = 14)

ax.text(50, 9, r'$\rho = \frac{m}{V}$', fontsize = 14)

ax.set_xlabel('x value', fontsize = 16)
ax.set_ylabel(r'$\Theta$', fontsize = 16)

ax.set_xlim(0, 100)
ax.set_ylim(0, 10)
```
