```python
import cartopy.crs as ccrs
import cartopy.feature as cfeat
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(1, 1, 1, projection = ccrs.LambertConformal())

ax.add_feature(cfeat.LAND)
ax.add_feature(cfeat.OCEAN)
ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.BORDERS, linestyle = ':')
ax.add_feature(cfeat.LAKES, alpha = 0.5)

state_borders = cfeat.NaturalEarthFeature(category = 'cultural', name = 'admin_1_states_provinces_lakes', scale = '50m', facecolor = 'none')

ax.add_feature(state_borders, linestyle = ':', edgecolor = 'black')

ax.set_extent([-130, -60, 20, 55])
```

```python
fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(1, 1, 1, projection = ccrs.LambertConformal())

ax.add_feature(cfeat.LAND)
ax.add_feature(cfeat.OCEAN)
ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.BORDERS, linestyle = ':')
ax.add_feature(cfeat.LAKES, alpha = 0.5)
ax.add_feature(cfeat.STATES.with_scale('50m'), linestyle = ':', edgecolor = 'black')

ax.set_extent([-95, -89, 35, 40])

ax.scatter(-94.2088, 36.3729, transform = ccrs.PlateCarree())
```
