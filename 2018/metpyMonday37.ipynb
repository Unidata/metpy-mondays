## 37 - Twin X-Axes

[Youtube](https://www.youtube.com/watch?v=lrREMkpHfao)

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(100)
y1 = np.random.random(100)
y2 = np.random.random(100) * 100
```

```python
plt.plot(x, y1)
```

```python
plt.plot(x, y2)
```

```python
fig = plt.figure(figsize = (10, 8))
ax = plt.subplot(1, 1, 1)
ax2 = ax.twinx()
ax.plot(x, y1)
ax2.plot(x, y2, color = 'tab:orange')

ax.set_ylabel('variable 1', fontsize = 16)
ax.yaxis.label.set_color('tab:blue')
ax.tick_params(axis = 'y', colors = 'tab:blue')
ax.spines['left'].set_color('tab:blue')

ax2.set_ylabel('variable 2', fontsize = 16)
ax2.yaxis.label.set_color('tab:orange')
ax2.tick_params(axis = 'y', colors = 'tab:orange')
ax2.spines['right'].set_color('tab:orange')
```
