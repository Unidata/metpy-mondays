```python
from netCDF4 import Dataset
```

```python
file = 'satellite_data/goes13.2011.117.200229.BAND_01.nc'
```

```python
ds = Dataset(file, 'r')
```

```python
list(ds.variables)
```

```python
ds.variables['data']
```

```python
vis_16bit = ds.variables['data'][0]
vis_16bit.max()
```

```python
vis_10bit = vis_16bit/32
vis_10bit.max()
```

```python
ds.variables['lat'][:]
```

```python
import numpy as np
```

```python
lon = np.ma.masked_values(ds.variables['lon'], 2.14329e+09)
lat = np.ma.masked_values(ds.variables['lat'], 2.14329e+09)
```

```python
print(lon)
```

```python
%matplotlib inline

import cartopy.crs as ccrs
import matplotlib.pyplot as plt

mapcrs = ccrs.Geostationary(central_longitude = -75)

fig = plt.figure(1, figsize = (19, 10))
ax = plt.subplot(111, projection = mapcrs)

img = ax.imshow(vis_10bit, cmap = 'Greys_r', origin = 'upper')
```

```python
# Values from Table 5. Visible-channel calibration coefficients for GOES-13 imager
m = 6.120196e-1
b = -17.749
k = 1.89544e-3
```

```python
vis_masked = np.ma.masked_array(vis_10bit, mask = lat.mask)
```

```python
# Equation 1
R = m * vis_masked

# Equation 5
eff_albedo = k * R

eff_albedo.max()
```

```python
fig = plt.figure(1, figsize = (19, 10))
ax = plt.subplot(111, projection = mapcrs)

img = ax.imshow(eff_albedo, cmap = 'Greys_r', origin = 'upper')
```

```python
eff_albedo_gamma_correction = np.sqrt(eff_albedo)
```

```python
fig = plt.figure(1, figsize = (19, 10))
ax = plt.subplot(111, projection = mapcrs)

img = ax.imshow(eff_albedo_gamma_correction, cmap = 'Greys_r', origin = 'upper')
```
