## 35 - matplotlib transforms

[Youtube](https://www.youtube.com/watch?v=CxsShxafkM8)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = range(10)
y = [i**2 for i in x]
```

```python
fig = plt.figure(figsize = (12,10))
ax = plt.subplot(1, 1, 1)
ax.plot(x, y, linestyle = 'None', marker = 'o')
# ax.text(0.5, 0.95, 'Top Center Label - My Plot', fontsize = 16, va = 'center', ha = 'center', transform = ax.transAxes)
ax.text(0.5, 0.95, 'Top Center Label - My Plot', fontsize = 16, va = 'center', ha = 'center', transform = fig.transFigure)
ax.set_xlim(-1, 100)
```
