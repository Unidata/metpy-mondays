```python
from siphon.catalog import TDSCatalog
```

```python
cat = TDSCatalog('https://thredds-test.unidata.ucar.edu/thredds/catalog/grib/NCEP/GFS/Global_0p5deg/catalog.xml')
```

```python
cat.datasets
```

```python
ds = cat.datasets['Full Collection (Reference / Forecast Time) Dataset']
```

```python
ncss = ds.subset()
```

```python
ncss.variables
```

```python
query = ncss.query()
query.accept('netcdf4')
query.variables('Temperature_surface')
query.lonlat_point(-97.330, 37.687)
data = ncss.get_data(query)
```

```python
list(data.variables)
```

```python
from metpy.units import units
from netCDF4 import num2date
```

```python
time = num2date(data['time'][:], data['time'].units, only_use_python_datetimes=True, only_use_cftime_datetimes=False)
```

```python
time
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data['Temperature_surface'][:].data)
```

```python
temperature = data['Temperature_surface'][:].data * units('kelvin')
temperature = temperature.to('degF')
```

```python
import matplotlib.dates as mdates
```

```python
fig, ax = plt.subplots()
ax.plot(time, temperature)
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_minor_locator(mdates.HourLocator([0, 6, 12, 18]))
ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(ax.xaxis.get_major_locator()))
```
