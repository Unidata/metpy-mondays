```python
from datetime import datetime
from metpy.units import units
from siphon.simplewebservice.wyoming import WyomingUpperAir
```

```python
date = datetime(2017, 9, 10, 6)
station = 'MFL'
```

```python
df = WyomingUpperAir.request_data(date, station)
```

```python
df.head()
```

```python
df.units
```

```python
ds = df.to_xarray()
ds
```

```python
for var_name, unit in df.units.items():
	if unit:
		ds[var_name] = ds[var_name] * units(unit)
ds
```

```python
ds = ds.set_coords(('pressure', 'height'))
ds
```

```python
ds.attrs = {'station': ds.isel(index=0)['station'].data, 
'station_number': ds.isel(index=0)['station_number'].data,
'latitude': ds.isel(index=0)['latitude'].data,
'longitude': ds.isel(index=0)['longitude'].data,
'elevation': ds.isel(index=0)['elevation'].data,
'pw': ds.isel(index=0)['pw'].data,
'time': datetime.strptime(str(ds.isel(index=0)['time'].data), '%Y-%m-%dT%H:%M:%S.000000000')}

ds
```

```python
ds = ds.drop(labels = ['station', 'station_number', 'time', 'latitude', 'longitude', 'elevation', 'pw'])
ds
```
