## 245 - Using a Fuzzy Search to Find Stations

[Youtube](https://www.youtube.com/watch?v=SPdlJ8EO-x0)

```python
from metpy.io import station_info
station_info.get('KDEN')

from fuzzywuzzy import fuzz
```

```python
score = fuzz.ratio('New York', 'Newyork')
score
```

```python
ids = [s for s in station_info]
ids[:10]
```

```python
ids = ids[7:]
ids[:10]
```

```python
search = 'Denver'
res = [station_info[stnid] for stnid in ids if fuzz.ratio(station_info[stnid].name, search) > 65]
res
```

```python
# search = 'springs'
```

```python
scores = [fuzz.ratio(station_info[stnid].name, search) for stnid in ids]
scores
```

```python
import numpy as np
```

```python
args = np.argsort(scores)
```

```python
for a in args[-10:]:
	print(ids[a], station_info[ids[a]].name)
```
