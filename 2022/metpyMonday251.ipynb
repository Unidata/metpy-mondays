```python
import metpy.calc as mpcalc
import matplotlib.pyplot as plt
import numpy as np
from metpy.units import units
```

```python
def plot_divergence(wind_speed, wind_direction):
 u, v = mpcalc.wind_components(wind_speed, wind_direction)
 div = mpcalc.divergence(u, v, dx = 10 * units.km, dy = 10 * units.km)
 fig, ax = plt.subplots()
 x, y = np.meshgrid(np.arange(5), np.arange(5))
 ax.quiver(x, y, u, v, zorder = 10)
 ax.set_xlim(-1, 5)
 ax.set_ylim(-1, 5)
 c = ax.pcolor(x, y, div, vmin = -5, vmax = 5, cmap = 'coolwarm')
 plt.colorbar(c)
```

```python
wind_speed = np.ones((5, 5)) * (20 * units('m/s'))
wind_direction = np.ones((5, 5)) * (180 * units.degrees)
plot_divergence(wind_speed, wind_direction)
```
