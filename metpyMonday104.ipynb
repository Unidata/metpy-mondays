```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

%matplotlib inline
```

```python
df = pd.DataFrame({'Time': pd.date_range(start = pd.datetime(2019, 11, 4), periods = 100, freq = 'H'), 'Temperature': np.random.random((100)) + 25, 'Humidity': np.random.random((100)) + 56 , 'Solar': np.random.random((100)) + 1100, 'Pressure': np.random.random((100)) + 1012})

df.head()
```

```python
def plot_data(df, time_col, *args):
 fig, axes = plt.subplots(nrows = len(args), sharex = True)

 if len(args) == 1:
  axes = [axes]

 for i, arg in enumerate(args):
  axes[i].plot(df[time_col], df[arg])
  axes[i].set_ylabel(arg)
 fig.align_labels()
```

```python
plot_data(df, 'Time', 'Pressure', 'Temperature', 'Humidity')
```
