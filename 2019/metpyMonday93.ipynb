```python
import pandas as pd
%matplotlib inline
```

```python
data_url = 'http://www.mesonet.org/data/basic/mesonet/mts/2019/08/17/20190817nrmn.mts'
```

```python
df = pd.read_csv(data_url, skiprows = 2, delim_whitespace = True)
df.head()
```

```python
df['TAIR']
```

```python
df.loc[0]
```

```python
df.iloc[0]
```

```python
df['TIME'] = pd.datetime(2019, 8, 17) + pd.TimedeltaIndex(df['TIME'], unit = 'm')
df.head()
```

```python
df[df['TIME'] == pd.datetime(2019, 8, 17, 16, 15)]
```

```python
df.set_index('TIME', inplace = True)
df.head()
```

```python
df.loc[pd.datetime(2019, 8, 17, 16, 15)]
```

```python
df.iloc[100]
```

```python
df.plot(y = ['TAIR', 'RELH'], use_index = True)
```
