```python
from datetime import datetime
import matplotlib.pyplot as plt
from metpy.units import pandas_dataframe_to_unit_arrays
from metpy.plots import Hodograph
from siphon.simplewebservice.wyoming import WyomingUpperAir
```

```python
df = WyomingUpperAir.request_data(datetime(2019, 1, 1, 12), 'TOP')
```

```python
sounding = pandas_dataframe_to_unit_arrays(df)
```

```python
sounding.keys()
```

```python
sounding['pressure']
```

```python
fig, ax = plt.subplots(1, 1, figsize = (6, 6))

h = Hodograph(ax, component_range = 120)
h.add_grid(increment = 20)
h.plot(sounding['u_wind'], sounding['v_wind'])
```

```python
sounding['u_wind']
```

```python
fig, ax = plt.subplots(1, 1, figsize = (6, 6))

h = Hodograph(ax, component_range = 120)
h.add_grid(increment = 20)
h.plot_colormapped(sounding['u_wind'], sounding['v_wind'], sounding['speed'])
h.ax.set_xlabel(sounding['u_wind'].units)
h.ax.set_ylabel(sounding['v_wind'].units)
```
