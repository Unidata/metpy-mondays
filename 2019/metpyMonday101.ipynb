```python
temperature_data  = np.random.random((100, 16)) * 25
number_of_stations = temperature_data.shape[1]
```

```python
plot_cols = 3
plot_rows = int(np.ceil(number_of_stations / plot_cols))

fig, axes = plt.subplots(nrows = plot_rows, ncols = plot_cols, sharex = True, sharey = True, figsize = (12, 12))

for i in range(plot_rows):
	for j in range(plot_cols):
		station_number = j * plot_rows + i
		if station_number < number_of_stations:
			axes[i, j].plot(temperature_data[:, station_number])
```
