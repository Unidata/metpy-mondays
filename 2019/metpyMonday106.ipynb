```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y = np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100))
```

```python
np.shape(x)
```

```python
x
```

```python
def ackley(x, y):
	""" Compute the Ackley function of coordinate pair(s) """
	return (-20 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) - np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) + np.e + 20)
```

```python
z = ackley(x, y)
```

```python
fig, ax = plt.subplots()
cs = ax.contour(x, y, z, 1000, cmap = 'plasma' )
ax.set_aspect('equal')
fig.colorbar(cs)
```
