## 102 - Using Dictionaries to Store Plotting Parameters

[Youtube](https://www.youtube.com/watch?v=hQ2otIzyATY)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
station_data = pd.DataFrame({'temperature': np.random.random((100,)) * 25, 'pressure': np.random.random((100,)) + 1012, 'humidity': np.random.random((100,)) + 65, 'solar radiation': np.random.random((100,)) + 1000})
```

```python
station_data
```

```python
plot_settings = {'pressure': {'color': 'black', 'linestyle': '-'}, 'temperature': {'color': 'tab:red'}, 'humidity': {'color': 'tab:green', 'linewidth': 2}}
```

```python
n_fields = station_data.shape[1]
fig, axes = plt.subplots(nrows = n_fields, figsize = (12, 4), sharex = True)

for ax_num, field in enumerate(station_data.columns):
 field_settings = plot_settings.get(field, {})
 color = field_settings.get('color', 'black')
 linestyle = field_settings.get('linestyle', '-')
 linewidth = field_settings.get('linewidth', 1)
 axes[ax_num].plot(station_data[field], color = color, linestyle = linestyle, linewidth = linewidth)
```
