## 98 - Pandas Merge/Join

[Youtube](https://www.youtube.com/watch?v=slUGaLyLJX0)


```python
import pandas as pd
```

```python
winds = pd.read_csv('wind_30min.txt')
temps = pd.read_csv('temps_hourly.txt')
```

```python
winds
```

```python
temps
```

```python
pd.merge(temps, winds)
```

```python
pd.merge(temps, winds, how = 'outer')
```

```python
pd.merge(temps, winds, how = 'left')
```

```python
pd.merge(temps, winds, how = 'right')
```

```python
winds.set_index('TIME', inplace = True)
temps.set_index('TIME', inplace = True)
```

```python
winds
```

```python
temps
```

```python
winds.join(temps)
```

```python
temps.join(winds)
```
