```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# From y = 3.45 * x + 2.12 + noise
x = np.array([0, 2, 3, 5, 6, 7, 9])
y = np.array([1.66, 8.29, 13.64, 18.45, 21.51, 26.77, 32.93])
```

```python
coeffs = np.polyfit(x, y, 1)
print(coeffs)
```

```python
x_fit = np.arange(10)
y_fit = np.polyval(coeffs, x_fit)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, label = 'Raw Data')
ax.plot(x_fit, y_fit, label = f'y = {coeffs[0]:.2f}x + {coeffs[1]:.2f}')
ax.set_xlabel('X Data', fontsize = 16)
ax.set_ylabel('Y Data', fontsize = 16)
ax.legend()
```
