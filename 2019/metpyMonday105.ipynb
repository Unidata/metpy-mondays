```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas.plotting import register_matplotlib_converters

%matplotlib inline
register_matplotlib_converters
```

```python
def make_plot(time, var, ylabel = '', **kwargs):
 # ylabel = kwargs.pop('ylabel', '')
 allowed_plot_kwargs = ['linewidth', 'color']
 # plot_kwargs = {k:kwargs.pop(k) for k, v in list(kwargs.items()) if k in allowed_plot_kwargs}
 plot_kwargs = {k:kwargs.pop(k) for k in list(kwargs) if k in allowed_plot_kwargs}
 fig, ax = plt.subplots()
 ax.plot(time, var, **plot_kwargs)
 print(kwargs)
 ax.set_ylabel(ylabel)
```

```python
t = pd.date_range(start = pd.datetime(2019, 11, 11), end = pd.datetime(2019, 11, 12), periods = 49)
temperature = np.random.random((len(t))) * 10 + 21
```

```python
make_plot(t, temperature, ylabel = 'Temperature', color = 'tab:red', linewidth = 0.5, linestyle = '--')
```
