```python
fig, axes = plt.subplots(2, 2, sharex = True, sharey = True)
for ax in axes.flat:
 data = 10 * np.random.randn(50, 50) + 15
 im = ax.imshow(data, cmap = 'plasma')
fig.colorbar(im, ax = axes[:, -1])
```
