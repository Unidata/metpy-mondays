```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.arange(256)
y = x + np.random.random((256)) * 10
```

```python
fig, ax = plt.subplots(figsize = (3.2, 3.2), dpi = 1100)
ax.plot(x, y, color = 'k')
fig.suptitle('Important Data')
ax.set_xlabel('The Independent Thing')
ax.set_ylabel('The Dependent Thing')
# plt.tight_layout()
plt.savefig('important_figure.png')
```
