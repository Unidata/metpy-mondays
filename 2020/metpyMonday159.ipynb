## 159 - Using Decorators to Mark Draft Plots


[Youtube](https://www.youtube.com/watch?v=QtWU09Im3ks)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def draft(*args, **kwargs):
	def wrapper(func):
		fig = plt.figure(**kwargs)
		fig.text(0.5, 0.5, 'Draft', fontsize = 65, color = 'gray', ha = 'center', va = 'center', alpha = 0.35, rotation = 25, transform = fig.transFigure)
		func(fig)
	return wrapper


def final(*args, **kwargs):
	def wrapper(func):
		fig = plt.figure(**kwargs)
		func(fig)
	return wrapper
```

```python
@draft(figsize = (8, 4))
def make_my_plot(fig):
	ax = plt.subplot()
	x = np.linspace(0, 2 * np.pi, 100)
	y = np.sin(x)
	ax.plot(x, y)
	return fig, ax
```

```python
@final(figsize = (8, 4))
def make_my_plot(fig):
	ax = plt.subplot()
	x = np.linspace(0, 2 * np.pi, 100)
	y = np.sin(x)
	ax.plot(x, y, color = 'tab:green')
	return fig, ax
```
