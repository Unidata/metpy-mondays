```python
from siphon.catalog import TDSCatalog
```

```python
cat = TDSCatalog('https://thredds-test.unidata.ucar.edu/thredds/catalog/noaaport/text/metar/catalog.xml')
cat.datasets
```

```python
ds = cat.datasets[0]
ds.download()
```

```python
from metpy.io import parse_metar_file
```

```python
ds.name
```

```python
df = parse_metar_file(ds.name)
df.head()
```

```python
df.units
```

```python
df_kmyj = df[df['station_id'] == 'KMYJ']
df_kmyj
```

```python
df_freezing = df[df['air_temperature'] <= 0]
df_freezing.head()
```

```python
print(f"{len(df_freezing['station_id'].unique())}/{len(df['station_id'].unique())} stations are below freezing.")
```
