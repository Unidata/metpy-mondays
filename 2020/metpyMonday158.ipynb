```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
time = pd.date_range(start = '11/23/20', end = '11/24/20', periods = 25)
time
```

```python
temps = []
for i in np.arange(20, 22, 0.1):
	temps.append(np.sin(np.arange(25) * 2 * 3.14 * 1/24) + i)
```

```python
for i in temps:
	plt.plot(time, i)
```

```python
cmap = matplotlib.cm.get_cmap('coolwarm')
norm = matplotlib.colors.Normalize(vmin = -1, vmax = 1)
```

```python
norm(0)
```

```python
cmap(0.5)
```

```python
cmap(norm(0))
```

```python
fig, ax = plt.subplots()
average = 21
for item in temps:
	ax.plot(time, item, color = cmap(norm(np.mean(item) - average)))
```

```python
from matplotlib.lines import Line2D
```

```python
legend_lines = [Line2D([0], [0], color = cmap(0), lw = 2), Line2D([0], [0], color = cmap(0.5), lw = 2), Line2D([0], [0], color = cmap(1.), lw = 2)]
```

```python
legend_labels = ['Cooler', 'Average', 'Warmer']
```

```python
fig, ax = plt.subplots(figsize = (12, 10))
average = 21
for item in temps:
	ax.plot(time, item, color = cmap(norm(np.mean(item) - average)))
ax.legend(legend_lines, legend_labels)
```
