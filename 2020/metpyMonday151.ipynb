```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
times = pd.date_range(start = '10/5/20', end = '10/10/20', freq = '1H')
```

```python
t1 = np.sin(1 / 24 * 2 * 3.14 * np.arange(121)) + (np.random.random((121)) * 0.5) + 25
t2 = np.sin(1 / 24 * 2 * 3.14 * np.arange(121) + 1) + (np.random.random((121)) * 0.5) + 24
```

```python
plt.plot(times, t1)
plt.plot(times, t2)
```

```python
fig, ax = plt.subplots(figsize = (10, 7.5))

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

plt.yticks(fontsize = 14)
plt.xticks(fontsize = 14)

ax.grid(linestyle = '--', alpha = 0.5)

l1, = ax.plot(times, t1)
l2, = ax.plot(times, t2)

ax.set_xlim(times[0], times[-1])

plt.tick_params(axis = 'both', which = 'both', bottom = False, top = False, labelbottom = True, left = False, right = False, labelleft = True)

ax.text(times[-1], t1[-1], 'KTUL', color = l1.get_color(), fontsize = 16)
ax.text(times[-1], t2[-1], 'KDEN', color = l2.get_color(), fontsize = 16)
```
