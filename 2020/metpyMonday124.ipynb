## 124 - Aligning Labels in Matplotlib

[Youtube](https://www.youtube.com/watch?v=KuseP7Fld-U)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(100)
y = np.random.random((100))
```

```python
fig, axes = plt.subplots(nrows = 2)
axes[0].plot(x, y)
axes[1].plot(x, y * 1000)

axes[0].set_ylabel('Variable 1')
axes[1].set_ylabel('Variable 2')

fig.align_labels()
```
