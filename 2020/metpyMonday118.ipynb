## 118 - Legends and Subplots

[Youtube](https://www.youtube.com/watch?v=wqsnZjkoOFI)

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x) + np.random.random(100) * 0.1
z = np.cos(x) + np.random.random(100) * 0.15
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 2)
l1, = ax1.plot(x, y, label = 'sin')
l2, = ax2.plot(x, z, label = 'cos', linestyle = ':', color = 'tab:orange')

fig.legend([l1, l2], ['sin function', 'cos function'], loc = (0.71, 0.82))
```
