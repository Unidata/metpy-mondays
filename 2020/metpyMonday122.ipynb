## 122 - Fitting Non-Linear Functions with a Linear Fit 
 
[Youtube](https://www.youtube.com/watch?v=K7MbYHvM-Rc)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
side_lengths = np.arange(1, 1000, 100)
areas = side_lengths * side_lengths
```

```python
plt.scatter(side_lengths, areas)
```

```python
plt.semilogx(side_lengths, areas)
```

```python
plt.semilogy(side_lengths, areas)
```

```python
plt.loglog(side_lengths, areas, linestyle = 'None', marker = 'o')
```

```python
coeffs = np.polyfit(np.log10(side_lengths), np.log10(areas), 1)
coeffs
```

```python
fit_y = np.polyval(coeffs, np.log10(side_lengths))
```

```python
plt.loglog(side_lengths, areas, linestyle = 'None', marker = 'o')
```

```python
plt.loglog(side_lengths, 10**fit_y)
```
