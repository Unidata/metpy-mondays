[Data](http://gis.ess.washington.edu/data/raster/thirtymeter/mtsthelens/)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im = plt.imread('./tifs/new10.tif')
im
```

```python
im.shape
```

```python
plt.imshow(im)
```

```python
topo_colors = [(31/255, 119/255, 180/255), (48/255, 147/255, 67/255), (140/255, 86/255, 75/255), (207/255, 207/255, 207/255)]
topo_colors
```

```python
from matplotlib.colors import LinearSegmentedColormap
```

```python
my_cmap = LinearSegmentedColormap.from_list('topo_basic', topo_colors, N = 10)
```

```python
plt.imshow(im, cmap = my_cmap, vmin = -1000, vmax = 3000)
plt.colorbar()
```
