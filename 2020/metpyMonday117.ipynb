```python
from metpy.cbook import get_test_data
import xarray as xr
```

```python
narr = xr.open_dataset(get_test_data('narr_example.nc', as_file_obj = False))
```

```python
from metpy.plots.declarative import *
from metpy.units import units
```

```python
narr
```

```python
import metpy.calc as mpcalc
```

```python
smoothed = mpcalc.smooth_n_point(narr['Temperature'].sel(isobaric = 850).isel(time = 0), 5, 3)
```

```python
narr['Temperature_5pt_3pass'] = (['y', 'x'], smoothed.data)
```

```python
narr['Temperature_5pt_3pass'].attrs = narr['Temperature'].attrs
```

```python
contour_raw = ContourPlot()
contour_raw.data = narr
contour_raw.field = 'Temperature'
contour_raw.level = 850 * units.hPa
contour_raw.linecolor = 'tab:red'
contour_raw.linewidth = 1
contour_raw.linestyle = '--'
contour_raw.contours = 15

contour_smooth = ContourPlot()
contour_smooth.data = narr
contour_smooth.field = 'Temperature_5pt_3pass'
contour_smooth.linecolor = 'tab:blue'
contour_smooth.linewidth = 1
contour_smooth.linestyle = '--'
contour_smooth.contours = 15

panel = MapPanel()
panel.area = 'co'
panel.layers = ['coastline', 'borders', 'states']
panel.title = 'Smoothing of 850hPa Temperatures'
panel.plots = [contour_raw, contour_smooth]

pc = PanelContainer()
pc.size = (10, 8)
pc.panels = [panel]
pc.show()
```
