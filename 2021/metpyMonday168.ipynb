```python
import numpy as np
import xarray as xr
import pandas as pd
```

```python
latitude = np.arange(20, 50.5, 0.5)
longitude = np.arange(-125, -65.5, 0.5)
print(np.shape(latitude), np.shape(longitude))
```

```python
levels = np.array([1000, 850, 500, 250])
print(np.shape(levels))
```

```python
times = pd.date_range(start = '2/1/21 00:00:00', end = '2/1/21 23:00:00', freq = 'H')
print(np.shape(times))
```

```python
temperature = np.random.random((24, 4, 61, 119))
print(np.shape(temperature))
```

```python
heights = np.array([0, 500, 1000, 2000])
print(np.shape(heights))
```

```python
wind_speed = np.random.random((24, 4, 61, 119))
print(np.shape(wind_speed))
```

```python
temperature_da = xr.DataArray(temperature, coords = (times, levels, latitude, longitude), dims = ['time', 'level', 'latitude', 'longitude'])
temperature_da.name = 'Air Temperature'
temperature_da.attrs['units'] = 'degC'
temperature_da
```

```python
wspd_da = xr.DataArray(wind_speed, coords = (times, levels, latitude, longitude), dims = ['time', 'level', 'latitude', 'longitude'])
wspd_da.name = 'Wind Speed'
wspd_da.attrs['units'] = 'knots'
wspd_da
```

```python
ds = xr.Dataset({'air_temperature': (['time', 'level', 'latitude', 'longitude'], temperature_da.data), 'wind_speed': (['time', 'height', 'latitude', 'longitude'], wspd_da.data)}, coords = {'time': times, 'level': levels, 'height': heights, 'latitude': latitude, 'longitude': longitude})
ds.attrs['title'] = 'My Custom Model'
ds
```

```python
single_time_temps = ds['air_temperature'].sel(time = '2021-02-01 13:00:00', level = 500)
single_time_temps
```

```python
np.shape(single_time_temps)
```

```python
single_time_temps.mean()
```
