```python
tracks = {name: geopandas.read_file(url) for name, url in storms.items()}
```

```python
tracks['Arthur']
```

```python
colors = ['blue', 'blueviolet', 'brown', 'burlywood', 'cadetblue', 'chartreuse', 'chocolate', 'coral', 'cornflowerblue' ,'crimson', 'darkblue', 'darkslategray', 'darkorange', 'tomato', 'violet' ,'wheat' ,'yellowgreen', 'springgreen', 'skyblue', 'sienna', 'darkgreen', 'darkkhaki', 'darkmagenta', 'darksalmon', 'deepskyblue', 'black', 'cyan', 'azure', 'blanchedalmond', 'seagreen', 'royalblue']
```

```python
geos = []

for i, t in enumerate(tracks):
	t = tracks[t]
	geo = PlotGeometry()
	geo.geometry = t['geometry']
	geo.fill = colors[i]
	geo.stroke = 'white'
	geos.append(geo)
	
panel = MapPanel()
panel.projection = ccrs.PlateCarree()
panel.layers = ['lakes', 'land', 'ocean', 'states']
panel.title = '2020 Hurricane Season'
panel.plots = geos

pc = PanelContainer()
pc.size = (15, 10)
pc.panels = [panel]

pc.show()
```
