```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def calculate_mean(values):
	sum_of_values = np.sum(values)
	return sum_of_values / len(values)
```

```python
def fit_line(x, y):
	mean_x = calculate_mean(x)
	mean_y = calculate_mean(y)
	slope = np.sum((x - mean_x) * (y - mean_y))
	slope /= np.sum((x - mean_x) ** 2)
	intercept = mean_y - slope * mean_x
	return [slope, intercept]
```

```python
x_data = [1, 2, 3, 4]
y_data = [1, 2, 3, 4]
plt.scatter(x_data, y_data)
```

```python
fit_line(x_data, y_data)
```

```python
np.polyfit(x_data, y_data, 1)
```
