```python
import itertools
```

```python
i = 0
for item in itertools.cycle('Metpy Forever'):
	print(item)
	i += 1
	if i > 50:
		break
```

```python
my_counter = itertools.count()
```

```python
i = 0
for index in my_counter:
	print(index)
	i += 1
	if i > 30:
		break
```

```python
temps = [[25, 26, 27, 28], [20, 20, 20, 21], [26, 20, 21, 22], [15, 16, 13, 11]]
```

```python
import matplotlib.pyplot as plt
```

```python
counter = itertools.count()
colors = itertools.cycle(['red', 'blue', 'green'])
```

```python
for data in temps:
	plot_num = next(counter)
	plt.figure()
	plt.plot(data, color = next(colors))
	plt.title(f'Data Plot {plot_num}')
```
