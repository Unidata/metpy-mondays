## 172​​ - Making Combinations with Itertools


[Youtube](https://www.youtube.com/watch?v=HuctnnGtk5c)

```python
import numpy as np
import itertools
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
temperature = [24, 25, 26, 27, 28]
dewpoint = [20, 19, 18, 17, 16]
pressure = [30.07, 30.08, 30.07, 30.06, 30.07]
windspeed = [15, 12, 15, 16, 17]
gust = [18, 18, 19, 20, 20]
wave_height = [2, 1.8, 2, 3, 4]
```

```python
data = {'temperature': temperature, 'dewpoint': dewpoint, 'pressure': pressure, 'windspeed': windspeed, 'gust': gust, 'wave_height': wave_height}
```

```python
list(itertools.permutations(data.keys(), 2))
```

```python
list(itertools.combinations_with_replacement(data.keys(), 2))
```

```python
list(itertools.combinations(data.keys(), 2))
```

```python
list(itertools.combinations(data.keys(), 3))
```

```python
for xvar, yvar in itertools.combinations(data.keys(), 2):
	fig, ax = plt.subplots()
	ax.scatter(data[xvar], data[yvar])
	ax.set_xlabel(xvar)
	ax.set_ylabel(yvar)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
for xvar, yvar, zvar in itertools.combinations(data.keys(), 3):
	fig = plt.figure()
	ax = fig.add_subplot(111, projection = '3d')
	ax.scatter(data[xvar], data[yvar], data[zvar])
	ax.set_xlabel(xvar)
	ax.set_ylabel(yvar)
	ax.set_zlabel(zvar)
```
