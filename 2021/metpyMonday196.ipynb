## 196 - Declarative Geometry Plotting - Ida Wind Probability

[Youtube](https://www.youtube.com/watch?v=sxBz1lyl7m8)

```python
import geopandas
from metpy.plots.declarative import PlotGeometry, MapPanel, PanelContainer
```

```python
wind_probs = geopandas.read_file('https://www.nhc.noaa.gov/gis/forecast/archive/2021082800_wsp_120hr5km.zip', engine = 'pyogrio')
wind_probs
```

```python
geo = PlotGeometry()
geo.geometry = wind_probs['geometry']
geo.fill = ['none', 'darkgreen', 'green', 'lime', 'gold', 'goldenrod', 'darkgoldenrod', 'orange', 'red', 'darkred', 'violet']

geo.stroke = 'none'
geo.labels = wind_probs['PERCENTAGE']
geo.label_facecolor = 'white'

panel = MapPanel()
panel.area = 'la'
panel.projection = 'lcc'
panel.layers = ['land', 'ocean', 'states']
panel.title = 'Hurricane Ida Wind Probability'
panel.plots = [geo]

pc = PanelContainer()
pc.panels = [panel]
pc.show()
```
