### Dataset

A collection of DataArray objects with aligned dimensions. Datasets have data variables, dimensions, coordinates, and attributes.

```python
data
```

### Variables/DataArrays

We don't interact much with variables directly - they have dimensions, data, and attributes which describe a single array. Variables cannot exist outside of a Dataset or DataArray.
DataArrays ar n-dimensional arrays with dimension names, coordinates, and attributes. These are what MetPy calculations are designed to work with!

```python
data['Temperature_isobaric']
```

## Dimensions

Dimensions are an axis in which all but one degree of freedom is fixed. Think of x, y, or height for example.

## Coordinates

Coordinates are like tick labels for a dimension, so a single dimension can have more than one coordinate. Think of a height that can be described as MSL, AGL, or pressure. All represent the same height, just with a different labeling.

```python
data['Temperature_isobaric'].dims
```

```python
data['Temperature_isobaric'].coords
```

The `*` indicates a dimensioned coordinate - or one that corresponds to a dimension of the DataArray or Dataset. A coordinate without a `*` is a non-dimensioned coordinate and is not indexed. They can be handy for aux. labels of something.
