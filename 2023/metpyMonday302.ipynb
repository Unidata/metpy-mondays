## 302 - Dictionaries of Functions and functools.partial

[Youtube](https://www.youtube.com/watch?v=Bvlt_CDdI94)

```python
import numpy as np

from functools import partial
```

```python
def moving_average(npts, data):
	"""
	Creates a moving average of npts wide.
	"""
	return np.convolve(np.ones((npts)), data) / npts
```

```python
data = np.random.random((1000))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data, label = 'Raw')
plt.plot(moving_average(10, data), label = 'Smoothed')
plt.legend()
```

```python
light_filter_width = 3
```

```python
moving_average(light_filter_width, data)
```

```python
filters = {'light': partial(moving_average, 3), 'medium': partial(moving_average, 10), 'harsh': partial(moving_average, 50)}
```

```python
filters['light']
```

```python
filters['light'](data)
```

```python
plt.plot(data)
plt.plot(filters['light'](data))
plt.plot(filters['medium'](data))
plt.plot(filters['harsh'](data))
```
