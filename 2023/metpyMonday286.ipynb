## 286 - Drawing Fronts and Boundaries with Path Effects

[Youtube](https://www.youtube.com/watch?v=VXUXqoLky5Q)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from metpy.plots import (ColdFront, WarmFront, ScallopedStroke, ColdFrontogenesis, ColdFrontolysis, Dryline, WarmFrontogenesis, WarmFrontolysis, OccludedFront, OccludedFrontogenesis, OccludedFrontolysis, RidgeAxis, Squall, StationaryFront, StationaryFrontogenesis, StationaryFrontolysis)
```

```python
effects = [ColdFront, WarmFront, ScallopedStroke, ColdFrontogenesis, ColdFrontolysis, Dryline, WarmFrontogenesis, WarmFrontolysis, OccludedFront, OccludedFrontogenesis, OccludedFrontolysis, RidgeAxis, Squall, StationaryFront, StationaryFrontogenesis, StationaryFrontolysis]
```

```python
x = np.array([0, 10])
y = np.array([1, 1])
```

```python
fig, ax = plt.subplots(figsize = (8, 8))
for i, effect in enumerate(effects):
	ax.plot(x, y * i, path_effects = [effect()])

_ = ax.set_yticks(np.arange(len(effects)))
_ = ax.set_yticklabels([e.__name__ for e in effects])
_ = ax.set_xticks([])
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
crs = ccrs.LambertConformal(central_longitude = -100, central_latitude = 45)
```

```python
fig = plt.figure(figsize = (17, 12))
ax = fig.add_subplot(1, 1, 1, projection = crs)

ax.set_extent([235, 290, 20, 55])
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth = 0.5)
ax.add_feature(cfeature.STATES, linewidth = 0.5)
ax.add_feature(cfeature.BORDERS, linewidth = 0.5)

ax.text(-100, 43, 'L', color = 'red', size = 30, ha = 'center', va = 'center', transform = ccrs.PlateCarree())
ax.plot([-100, -99, -99, -100, -101], [42, 40, 36, 33, 30], path_effects = [ColdFront()], transform = ccrs.PlateCarree())
ax.plot([-99, -94, -90, -85, -80], [43, 43.5, 43, 40, 37], path_effects = [WarmFront()], transform = ccrs.PlateCarree())
```
