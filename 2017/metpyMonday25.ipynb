```python
import numpy as np
import matplotlib.pyplot as plt
from metpy.plots import ctables
%matplotlib inline
```

```python
x, y = np.mgrid[-5: 5: 0.05, -5: 5: 0.05]
x
```

```python
z = 5 * (np.sqrt(x**2 + y**2) + np.sin(x**2 + y**2))
```

```python
plt.plot(x[:, 100], z[:, 100])
```

```python
np.max(z)
```

```python
fig = plt.figure()
ax1 = plt.subplot(1, 1, 1)
ax1.imshow(z)
```

```python
fig = plt.figure()
ax1 = plt.subplot(1, 1, 1)

norm, cmap = ctables.registry.get_with_range('WVCIMSS_r', 0, 40)
ax1.imshow(z, norm = norm, cmap = cmap)
```
